```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data={
    'Name':['Danish','Shadman','Aditya','Shriyasha','Soumyashree','Abdullah','Fasahat','Ananya','Basit','Adnan','Anouk','Vishnu','Sneha'],
    'Age':[19,21,23,18,20,22,25,17,15,16,22,28,21],
    'Gender':['Male','Male','Male','Female','Female','Male','Male','Female','Male','Male','Female','Male','Female']
}
df=pd.DataFrame(data)
print(df)
```

```python
gender_counts=df['Gender'].value_counts()
gender_counts.plot(kind='bar',color=['Lightblue','pink'],zorder=2,edgecolor='black')
plt.title("Gender Distribution Graph",fontweight='bold',color='blue',fontsize=14)
plt.xlabel('Genders',fontweight='bold',fontsize=12)
plt.ylabel('Counts',fontweight='bold',fontsize=12)
plt.xticks(rotation=0,fontweight='bold')
plt.grid(True,zorder=0,linestyle='--',linewidth=1)
plt.show()
```
